```python
import matplotlib.pyplot as plt
import math as m
plt.figure(figsize=(20,10))
y1 = [x + 1 for x in range(1, 100)]
plt.plot(y1, c= 'brown', alpha = .5, label = 'O(1)')

y2 = [m.log(x) for x in range(1, 100)]
plt.plot(y2, c= 'blue', label = 'O(log n)')

y3 = [x for x in range(1, 100)]
plt.plot(y3, c= 'red', label = 'O(n)')

y4 = [x * m.log(x) for x in range(1,100)]
plt.plot(y4, c= 'purple', label = 'O(n log n)')

y5 = [pow(x, 2) for x in range(1, 100)]
plt.plot(y5, c = 'orange', label = 'O(n^2)')

y6 = [pow(2, x) for x in range(1, 100)]
plt.plot(y6, c = 'teal', label = 'O(2^n)')

y7 = [m.factorial(x) for x in range(1, 100)]
plt.plot(y7, c = 'black', label = 'O(n!)')

y8 = [pow(x,x) for x in range(1, 100)]
plt.plot(y8, c = 'pink', label = 'O(n^n)')

plt.xlim(1, 100)
plt.ylim(1, 200)
plt.title('Big O Notation')
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.legend()
plt.show()
```
