1. Plot the equation: y = x^2 - x + 2, add an annotation (origin) at (0,0)

```python
import matplotlib.pyplot as plt
numbers = list(range (100))
y = [x**2 - x + 2 for x in numbers] 
plt.plot(numbers, y)
plt.annotate('Origin', xy = (0,0), xytext = (0,250), fontsize = 10, color = 'red')
plt.show()
```

2. Create and label four equations: y = sqrt(x), y = x^3, y = 2^x, y = 1/(x+1) (can use math)

```python
import math as m
numbers = list(range(50))
y = [m.sqrt(x) for x in numbers]
plt.plot(numbers, y)
plt.title('Y = sqrt(X)')
plt.xlabel('X')
plt.ylabel('Sqrt(X)')
plt.show()
```

```python
numbers = list(range(10))
y = [pow(x,3) for x in numbers]
plt.plot(numbers, y)
plt.title('Y = X^3')
plt.xlabel('X')
plt.ylabel('X^3')
plt.show()
```

```python
numbers = list(range(10))
y = [pow(2, x) for x in numbers]
plt.plot(numbers, y)
plt.title('Y = 2^X')
plt.xlabel('X')
plt.ylabel('2^X')
plt.show()
```

```python
numbers = list(range(5))
y = [1/(x+1) for x in numbers]
plt.plot(numbers, y)
plt.title('Y = 1/(X+1)')
plt.xlabel('X')
plt.ylabel('1/X+1')
plt.show()
```

3. Combine the last four functions in to one large subplot.

```python
n_rows = 2
n_cols = 2
n1 = list(range(10))
n2 = list(range(50))

y = [m.sqrt(x) for x in n2]
plt.subplot(221)
plt.plot(n2, y)
plt.xlabel('X')
plt.ylabel('Sqrt(X)')

y = [pow(x,3) for x in n1]
plt.subplot(222)
plt.plot(n1, y)
plt.xlabel('X')
plt.ylabel('X^3')

y = [pow(2, x) for x in n1]
plt.subplot(223)
plt.plot(n1, y)
plt.xlabel('X')
plt.ylabel('2^X')

y = [1/(x+1) for x in n1]
plt.subplot(224)
plt.plot(n1, y)
plt.xlabel('X')
plt.ylabel('1/X+1')
plt.subplots_adjust(wspace=.5, hspace = .5)
plt.suptitle('Y = sqrt(X), Y = X^3, Y= 2^X, Y = 1/(X+1)')
plt.show()
```

4. Combine all four of the above in to one figure, with appropriate title and legend. 

```python
import math as m
import matplotlib.pyplot as plt
y1 = [m.sqrt(x) for x in range(4)]
y2 = [pow(x, 3) for x in range(4)]
y3 = [pow(2, x) for x in range(4)]
y4 = [(1/(x+1)) for x in range(4)]
plt.plot(y1, c= 'brown', alpha = .5, label = '$sqrt(x)$')
plt.plot(y2, c= 'red', alpha = .5, label = '$x^3$')
plt.plot(y3, c= 'orange', alpha = .5, label = '$2^x$')
plt.plot(y4, c= 'blue', alpha = .5, label = '$1/x+1$')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Various Interactions With X')
plt.legend()
plt.show()
```
